```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import userdata
```

```python
ruta = userdata.get('ruta')

df = pd.read_csv(ruta + 'wine_quality_red-clean.csv', index_col=0)

df.head()
```

```python
correlaciones = df.corr()
correlaciones
```

```python
def calcular_correlacion_pearson(df, columna_x, columna_y):
    # Verificar si las columnas existen en el DataFrame
    if columna_x not in df.columns or columna_y not in df.columns:
        raise ValueError(f"Las columnas '{columna_x}' y/o '{columna_y}' no existen en el DataFrame.")

    # Calcular la correlación de Pearson
    correlacion = df[columna_x].corr(df[columna_y], method='pearson')

    print(f"Coeficiente de correlación: {correlacion:,.5f}")
```

```python
# Vamos a generar una funcion generica, para evitar repeticion de codigo:
def graficar_dispersion(df, independiente, dependiente, titulo, etiqueta_x, etiqueta_y):
    plt.figure(figsize=(8, 6))  # Tamaño de la figura
    sns.scatterplot(x=independiente, y=dependiente, data=df, color='b', alpha=0.7, s=50)  # Gráfica de dispersión

    # Configuración de etiquetas y título
    plt.xlabel(etiqueta_x)
    plt.ylabel(etiqueta_y)
    plt.title(titulo)

    # Cuadrícula para facilitar la lectura
    plt.grid(True, linestyle='--', alpha=0.5)

    # Mostrar la gráfica
    plt.show()
```

```python
independiente='free_sulfur_dioxide'
dependiente='total_sulfur_dioxide'
titulo="Relación entre free_sulfur_dioxide y total_sulfur_dioxide"
etiqueta_x='Nivel de free_sulfur_dioxide'
etiqueta_y='Nivel de total_sulfur_dioxide'
```

```python
graficar_dispersion(df, independiente, dependiente, titulo, etiqueta_x, etiqueta_y)
```

```python
calcular_correlacion_pearson(df, independiente, dependiente)
```

```python
# Método IQR para eliminar valores atípicos
Q1 = df[['free_sulfur_dioxide', 'total_sulfur_dioxide']].quantile(0.25)
Q3 = df[['free_sulfur_dioxide', 'total_sulfur_dioxide']].quantile(0.75)
IQR = Q3 - Q1

# Definir límites para considerar datos válidos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar el dataframe sin valores atípicos
df_filtered = df[
    (df['free_sulfur_dioxide'] >= lower_bound['free_sulfur_dioxide']) &
    (df['free_sulfur_dioxide'] <= upper_bound['free_sulfur_dioxide']) &
    (df['total_sulfur_dioxide'] >= lower_bound['total_sulfur_dioxide']) &
    (df['total_sulfur_dioxide'] <= upper_bound['total_sulfur_dioxide'])
]

df_filtered.head()
```

```python
graficar_dispersion(df_filtered, independiente, dependiente, titulo, etiqueta_x, etiqueta_y)
```

```python
calcular_correlacion_pearson(df_filtered, independiente, dependiente)
```

```python
independiente='fixed_acidity'
dependiente='citric_acid'
titulo="Relación entre fixed_acidity y citric_acid"
etiqueta_x='Nivel de fixed_acidity'
etiqueta_y='Nivel de citric_acid'
graficar_dispersion(df, independiente, dependiente, titulo, etiqueta_x, etiqueta_y)
```

```python
calcular_correlacion_pearson(df, independiente, dependiente)
```

```python
filtro_fixed_acidity = (df['fixed_acidity'] > 0) & (df['fixed_acidity'] < 14)
filtro_citric_acid = df['citric_acid'] < 1

df_filtered = df[filtro_fixed_acidity & filtro_citric_acid]

df_filtered.head()
```

```python
graficar_dispersion(df_filtered, independiente, dependiente, titulo, etiqueta_x, etiqueta_y)
```

```python
calcular_correlacion_pearson(df_filtered, independiente, dependiente)
```
